```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mandelbrot(c, max_iter):
    z=0
    for i in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return max_iter
```

```python
#define the range of points in the complex plane

x_values = np.linspace(-2, 0.5, 1000)
y_values = np.linspace(-1.5, 1.5, 1000)

#choose max number of iterations
max_iter = 100
```

```python
#compute mandelbrot

mandelbrot_set = np.zeros((len(x_values), len(y_values)))

for i,x in enumerate(x_values):
    for j,y in enumerate(y_values):
        c = complex(x,y)
        mandelbrot_set[i,j] = mandelbrot(c, max_iter)
```

```python
#plot mandelbrot

plt.figure(figsize=(10,6))
plt.imshow(mandelbrot_set.T, extent=(-2, 0.5, -1.5, 1.5), cmap='winter', origin='lower', aspect='auto')
plt.colorbar()
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Mandelbrot Set')
plt.gca().set_asp("equal")
plt.show()
```
